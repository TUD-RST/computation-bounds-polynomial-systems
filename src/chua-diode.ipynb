# Characteristics of Chua's Diode

```python
%display latex
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['mathtext.fontset'] = 'cm'
```

```python
# piecewise linear characteristics
def diodelin(x1):
    m0 = -0.5
    m1 = -0.8
    i = m0*x1 +(1/2)*(m1-m0)*(abs(x1+1)-abs(x1-1))
    return i
```

```python
# cubic characteristics
def diodekub(x1):
    a = -4183/5120
    b = 189/4096
    i = a*x1 + b*x1^3
    return i
```

```python
# visualization
N = 100
u = np.linspace(-3,3,N)
ilin = np.zeros(N)
ikub = np.zeros(N)
for j,val in enumerate(u):
    ilin[j] = diodelin(val)
    ikub[j] = diodekub(val)
plt.plot(u,ilin, label="piecewise-linear",)
plt.plot(u,ikub, label="cubic")
plt.xlabel('$x_1$',fontsize=15)
plt.ylabel('$\\phi(x_1)$',fontsize=15)
plt.title('Characteristics of the Chua diode models')
plt.legend(fontsize=13)
plt.grid()
#plt.savefig('chua-diode.png', bbox_inches='tight')
plt.show()
plt.close()
```
